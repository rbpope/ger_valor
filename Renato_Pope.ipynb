# Azul Linhas Aéreas

```python
df = pd.read_excel('azul.xls')
```

```python
pme = abs((df['Estoques']/df['Custo de Bens e/ou Serviços Vendidos'])*90)
pmr = abs((df['Contas a Receber']/df['Receita Líquida de Vendas e/ou Serviços'])*90)
pmp = abs((df['Fornecedores']/df['Custo de Bens e/ou Serviços Vendidos'])*90)
co = pme + pmr
cc = co - pmp
```

```python
colunas = ['PME', 'PMR', 'PMP', 'CO', 'CC']
azul = pd.DataFrame(columns = colunas)
azul['PME'] = pme
azul['PMR'] = pmr
azul['PMP'] = pmp
azul['CO'] = co
azul['CC'] = cc
azul
```

```python
def ke(rf, rm, beta):
  rf = float(rf)
  rm = float(rm)
  beta = float(beta)
  ke = rf + beta*(rm - rf)
  return ke

def kd(i,t):
  i = float(i)
  t = float(t)
  kd = i * (1-t)
  return kd

def wacc(we,ke,wd,kd):
  we = float(we)
  wd = float(wd)
  wacc = (we + ke) * (kd + wd)
  return wacc
```

```python
taxa = wacc(0.5,ke(0.1,0.12,2),0.5,kd(0.1,0.34))
taxa
```

## Índices de Liquidez

```python
#Índice de Liquidez imediata
li = df['Caixa e Equivalentes de Caixa']/df['Passivo Circulante']
#Índice de Liquidez Corrente
lc = df['Ativo Circulante']/df['Passivo Circulante']
#Índice de Liquidez seca
ls = (df['Ativo Circulante'] - df['Estoques'])/df['Passivo Circulante']
#Calcular o Capital de Giro:
cg = df['Ativo Circulante']  - df['Passivo Circulante']
```

```python
colunas = ['LI', 'LC', 'LS', 'Capital de Giro']
liq = pd.DataFrame(columns = colunas)
liq['LI'] = li
liq['LC'] = lc
liq['LS'] = ls
liq['Capital de Giro'] = cg
liq
```

## Endividamento

```python
#Grau de endividamento de Curto Prazo
ecp = df['Passivo Circulante']/df['Ativo Total']
#Grau de endividamento de longo Prazo
elp = df['Passivo Não Circulante']/df['Ativo Total']
#Grau de endividamento total
et = (df['Passivo Circulante'] + df['Passivo Não Circulante'])/df['Ativo Total']
```

```python
colunas = ['ECP', 'ELP', 'ET']
endiv = pd.DataFrame(columns = colunas)
endiv['ECP'] = ecp
endiv['ELP'] = elp
endiv['ET'] = et
endiv
```

## Rentabilidade

```python
#ROE - Retorno Sobre o Equity
roe = df['Lucro/Prejuízo do Período']/df['Patrimônio Líquido']
#ROA - Lucro liquido sobre os ativos
roa = df['Lucro/Prejuízo do Período']/df['Ativo Total']
```

```python
colunas = ['ROE', 'ROA']
rentabilidade = pd.DataFrame(columns = colunas)
rentabilidade['ROE'] = roe
rentabilidade['ROA'] = roa
rentabilidade
```
